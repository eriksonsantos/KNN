```python
import random

amostras= np.array(random.sample(range(XC.shape[0]),XC.shape[0]))
```

```python
qtdTrain = int(XC.shape[0]*0.75)

trainX = tf.Variable(XC[amostras[0:qtdTrain]]);
trainY = tf.Variable(YC[amostras[0:qtdTrain]]);

testX = tf.Variable(XC[amostras[qtdTrain:(amostras.shape[0])]])
testY = tf.Variable(YC[amostras[qtdTrain:(amostras.shape[0])]])
```

```python
def getDist(p,xTrain,xTest):
    test = tf.Variable([1.,2.], dtype='float64')
    dist = np.zeros(shape=(xTest.shape[0], xTrain.shape[0]))
    

    for i in range(xTest.shape[0]):
        for j in range(xTrain.shape[0]): 
            dist[i,j] = np.power((tf.subtract( xTrain[j,0] ,xTest[i,0])),p) + np.power((tf.subtract( xTrain[j,1] ,xTest[i,1])),p)
            dist[i,j] = np.abs( dist[i,j])**(1/p)  
            
    return dist
```

```python
dist = getDist(3,trainX,testX)
```

```python
import pandas as pd

pd.DataFrame(dist)
```

```python
def getArgKminValues(dist,k):
    aux = dist
    arg = np.zeros(shape=(aux.shape[0],k), dtype='int32')
    for i in range(dist.shape[0]):
        arg[i,] = (dist[i].argsort()[0:k])
        
    return arg
```

```python
pd.DataFrame(getArgKminValues(dist,15))
```

```python
minPosic = getArgKminValues(dist,15)
```

```python
pd.DataFrame(trainX.numpy())
```

```python
pd.DataFrame(trainY.numpy()[minPosic])
```

```python
classAux = trainY.numpy()[minPosic]

import statistics
```

```python
def getModa(classAux):
    
    output = np.zeros(classAux.shape[0],dtype='int32')
    for i in range(classAux.shape[0]):
        output[i] = statistics.multimode(classAux[i])[0]
    return output
    
```

```python
classTest = getModa(classAux)


def defineColors(classTest):
    classTest_color = []
    for i in range(classTest.shape[0]):
        if classTest[i] == 1:
            classTest_color.append('red')
        if classTest[i] == 2:
            classTest_color.append('blue')
        if classTest[i] == 3:
            classTest_color.append('yellow')
        if classTest[i] == 4:
            classTest_color.append('green')
    return classTest_color
            
```

```python
def  createAmostras(sd):
    xc1 = np.random.normal(loc=(1,2),size=(100,2),scale=sd)
    xc2= np.random.normal(loc=(2,3),size=(100,2),scale=sd)
    xc3= np.random.normal(loc=(3,2),size=(100,2),scale=sd)
    xc4= np.random.normal(loc=(4,3),size=(100,2),scale=sd)
    
    XC = np.concatenate([xc1,xc2,xc3,xc4])
    
    qtdTrain = int(XC.shape[0]*0.75)

    trainX = tf.Variable(XC[amostras[0:qtdTrain]]);
    trainY = tf.Variable(YC[amostras[0:qtdTrain]]);

    testX = tf.Variable(XC[amostras[qtdTrain:(amostras.shape[0])]])
    testY = tf.Variable(YC[amostras[qtdTrain:(amostras.shape[0])]])
    
    return trainX,trainY,testX,testY
    
```

```python
Vetoracuracia = np.zeros(shape=(3,3))
```

```python
Vetoracuracia
```

```python
p = 3
for i in range(3): #K
    for j in range(3): #sd
        
        [trainX,trainY,testX,testY] = createAmostras(1/(j+3))
        distAux = getDist(p,trainX,testX)
        
        minPosic = getArgKminValues(distAux,(i+3))
        
        classAux = trainY.numpy()[minPosic]
        classTest = getModa(classAux)
        
        Vetoracuracia[i,j] = accuracy_score(testY,classTest)
        
        
```
